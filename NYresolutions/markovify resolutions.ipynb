```python
import markovify

with open('./NYresolutions.txt','r', encoding="utf8") as f:
    resolutions = f.readlines()

resolution_model = markovify.NewlineText(resolutions, state_size = 2)
for i in range(100):
    newline = resolution_model.make_sentence(max_overlap_ratio=0.6)
    if newline != None:
        print(newline)
        print(' ')
```
